```python
order_review_df[order_review_df.review_comment_title.isna()]
```

```python
order_review_df.review_comment_title.value_counts()
```

```python
order_review_df["review_comment_title"] = order_review_df["review_comment_title"].str.title()
```

```python
order_review_df.review_comment_title.value_counts()
```

```python
def fill_review_title(row):
    if pd.isna(row["review_comment_title"]):
        if row["review_score"] == 5:
            return "Recomendo"
        elif row["review_score"] == 4:
            return "Muito Bom"
        elif row["review_score"] == 3:
            return "Bom"
        elif row["review_score"] == 2:
            return "Gerou Dúvida"
        elif row["review_score"] == 1:
            return "Irritante"
    return row["review_comment_title"]

order_review_df["review_comment_title"] = order_review_df.apply(fill_review_title, axis=1)
```

```python
order_review_df.isna().sum()
```

```python
order_review_df.review_comment_message.value_counts()
```

```python
order_review_df["review_comment_message"] = order_review_df["review_comment_message"].str.title()
```

```python
order_review_df.review_comment_message.value_counts()
```

```python
def fill_review_message(row):
    if pd.isna(row["review_comment_message"]):
        if row["review_score"] == 5:
            return "Ótimo! Produto excelente, recomendo."
        elif row["review_score"] == 4:
            return "Muito bom! Atendeu minhas expectativas."
        elif row["review_score"] == 3:
            return "Bom, mas poderia ser melhor."
        elif row["review_score"] == 2:
            return "Não gostei muito, esperava mais."
        elif row["review_score"] == 1:
            return "Péssimo! Não recomendo."
    return row["review_comment_message"]

order_review_df["review_comment_message"] = order_review_df.apply(fill_review_message, axis=1)
```

```python
order_review_df = order_review_df.dropna(subset=['review_comment_message'])
```

### Pertanyaan 4 : Recency (R) - Kapan terakhir kali pelanggan melakukan pembelian?
### Pertanyaan 5 : Frequency (F) - Seberapa sering pelanggan melakukan pembelian?
### Pertanyaan 6 : Monetary (M) - Berapa banyak uang yang dihabiskan pelanggan?

```python
# RFM Analysis
rfm_df = all_data.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max",  
    "order_id": "nunique",  
    "price": "sum"  
})

rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = all_data["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

rfm_df.head()
```

```python
rfm_df['short_customer_id'] = rfm_df['customer_unique_id'].astype(str).str[:5]

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(40, 8))

sns.barplot(y="recency", x="short_customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), ax=ax[0])
sns.barplot(y="frequency", x="short_customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), ax=ax[1])
sns.barplot(y="monetary", x="short_customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5),ax=ax[2])

# Menambahkan rotasi label X
ax[0].tick_params(axis='x', labelsize=20, rotation=45)
ax[1].tick_params(axis='x', labelsize=20, rotation=45)
ax[2].tick_params(axis='x', labelsize=20, rotation=45)

# Menambahkan judul
plt.suptitle("Best Customer Based on RFM Parameters (customer_unique_id)", fontsize=25)

plt.show()
```

### Pertanyaan 7:seberapa akurat estimasi pengiriman (order_estimated_delivery_date) dibandingkan dengan tanggal sebenarnya

```python
all_data["delivery_diff"] = (
    all_data["order_delivered_customer_date"] - all_data["order_estimated_delivery_date"]
).dt.days
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(all_data["delivery_diff"], bins=30, kde=True, color="blue")

plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Estimasi tepat waktu")
plt.xlabel("Selisih (Hari) antara Estimasi dan Realisasi")
plt.ylabel("Jumlah Pesanan")
plt.title("Distribusi Akurasi Estimasi Pengiriman")
plt.legend()
plt.show()
```

```python
# Clustering
all_data["Delivery_Category"] = pd.cut(
    all_data["delivery_diff"],
    bins=[-float("inf"), -1, 0, 7, float("inf")],  
    labels=["Early", "On Time", "Late", "Very Late"]
)

# Cek hasilnya
all_data[["order_delivered_customer_date", "order_estimated_delivery_date", "delivery_diff", "Delivery_Category"]].head()
```

```python
plt.figure(figsize=(9, 5))

sns.countplot(
    data=all_data, 
    x="Delivery_Category", 
    hue="Delivery_Category", 
    palette={"Early": "green", "On Time": "blue", "Late": "orange", "Very Late": "red"},
    legend=False  
)

plt.title("Distribusi Kategori Pengiriman Berdasarkan Ketepatan Waktu")
plt.xlabel("Kategori Pengiriman")
plt.ylabel("Jumlah Order")
plt.show()
```

```python
Bagaimana distribusi geografis pelanggan dan penjual berdasarkan kota atau negara bagian?
```

```python
gdf = gpd.GeoDataFrame(all_data, 
                       geometry=gpd.points_from_xy(all_data['geolocation_lng'], all_data['geolocation_lat']),
                       crs="EPSG:4326")

# Plot peta distribusi pelanggan
world = gpd.read_file("data/ne_110m_admin_0_countries.shp") 
fig, ax = plt.subplots(figsize=(10, 6))
world[world.name == "Brazil"].plot(ax=ax, color="lightgray")
gdf.plot(ax=ax, markersize=5, color="red", alpha=0.5)
plt.title("Distribusi Pelanggan berdasarkan Lokasi")
plt.show()
```
